```python
#import csv
hollywood_df = pd.read_csv('Hollywood.csv', encoding = "latin")
hollywood_df.head()
```

```python
#import 
bollywood_df = pd.read_csv('Bollywood.csv', encoding = "latin")
bollywood_df.head()
```
