```python
import pandas as pd
import numpy as np
import scipy as sp
import datetime as dt
import matplotlib as plt
```

```python
#import csv
hollywood_df = pd.read_csv('Hollywood.csv', encoding = "latin")
hollywood_df.head()
```

```python
#import 
bollywood_df = pd.read_csv('Bollywood.csv', encoding = "latin")
bollywood_df.head()
```

```python
#import 
tv_df = pd.read_csv('TV.csv', encoding = "latin")
tv_df.head()
```

```python
#combine dataframes
movies_df = pd.concat([hollywood_df, bollywood_df, tv_df], ignore_index=True)
alpha_df = movies_df.sort_values(by=['Title'])
alpha_df.head()
```

```python
alpha_df.groupby('Title Type').plot.bar(figsize=(8,8), color='red', alpha=0.5)
#plt.xlabel('Source', size=12)
#plt.ylabel('Number of Anime', size=12)
#plt.title('Number of Anime per Source', size=20)
plt.tight_layout()
#plt.savefig('WebVisualizations/Images/numberofanimepersource.png', bboxinches = 'tight')
```
